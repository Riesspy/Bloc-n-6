# Explore cluster created by our model

```python
import os
import geopandas as gpd
```

```python
cluster_points = "../prod/model/cluster_points.geojson"
gdf = gpd.read_file(cluster_points)

gdf["cluster"].unique()
```

```python
gdf.shape
```

```python
mask_little = gdf["cluster"].isin([-1, 0, 1, 2, 4, 5, 6, 8, 12, 15])
gdf_mask = gdf[~mask_little]

gdf_mask.groupby(["cluster"]).count()
```
